```python
import numpy as np
import librosa as lb
import os, shutil
import soundfile as sf

from src.projects.voice_enhancement.fourier_transform import filter_downsample
```

```python
original_path = r"C:\Datasets\Clean and Noisy Audio Dataset"
target_path = r"C:\Datasets\Clean and Noisy Audio Dataset 16kHz"
sr_target = 16000 # we want 16kHz for LSTM (downsample from 48kHz)

for split_orig, split_target in [('trainset_56spk', 'train'), ('testset', 'test')]:
    for cls in ['clean', 'noisy']:
        src_dir = os.path.join(original_path, f"{cls}_{split_orig}_wav")
        dest_split_dir = os.path.join(target_path, split_target)
        os.makedirs(dest_split_dir, exist_ok=True)
        dest_cls_dir = os.path.join(dest_split_dir, cls)
        os.makedirs(dest_cls_dir, exist_ok=True)
        
        print(f"Processing: {src_dir} -> {dest_cls_dir}")
        
        for filename in os.listdir(src_dir):
            if filename.endswith('wav'):
                src_file_path = os.path.join(src_dir, filename)
                dst_file_path = os.path.join(dest_cls_dir, filename)
                
                audio, sr_orig = lb.load(src_file_path, sr=None) # keep original SR
                # audio_16k = filter_downsample(audio, sr_orig, sr_target) we would need terabytes of RAM to run our version
                audio_16k = lb.resample(audio, orig_sr=sr_orig, target_sr=sr_target)
                
                sf.write(dst_file_path, audio_16k, sr_target)
                
        print("Part complete.")
```
